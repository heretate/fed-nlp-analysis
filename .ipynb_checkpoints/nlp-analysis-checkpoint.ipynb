# Analyzing FOMC Meetings with NLP

In the context of the rising inflation and looming recession, the Fed's policy decisions has become a focal point of scrutiny. Being able to effectively parse and analyze the discussions during the FOMC meetings have become increasingly important. Here, we present one approach to modelling the FOMC's meeting using NLP for the purpose of deriving sentiment analysis as well as prediction on future guidance.

## Plan

The goal is to be able to reliably classify sentiment given the Fed meeting data. To do so, we will go through a few stages: Data pre-processing, exploratory analysis, and finally modelling. We will first look at data pre-processing.

## Data Pre-processing

Here, we will source our data from https://www.federalreserve.gov/monetarypolicy/fomccalendars.htm

Supplementary sources will be considered and added if they can help improve model performance.

FOMC data comes through in a couple sources: 

- Statement
- Press Conference
- Minutes

The statement, minutes, and press conference transcripts are all available via pdf. We will aim to parse these documents into something usable for our model. Our target time period is starting 2019.

```python
import pickle
import datetime as dt
import os
import re
import pandas as pd
import wget
import unzip
import numpy as np
import torch
import torch.nn as nn
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

import keras
from keras.preprocessing.text import Tokenizer
from keras_preprocessing.sequence import pad_sequences
from keras.layers import Dense, Embedding, LSTM, SpatialDropout1D
from keras.models import Sequential


# Custom modules
import data_extraction_utils as deu
import data_visualization as dv
```

```python
# Using beautiful soup, we first search the calendar for all possible text references 
# to monetary press release and extract the date.

run_extraction = False  # Runs the entire extraction process 
pickle_file_name = 'data/fomc_data_2022-07-17.pkl'  # If we are not running extraction, provide file name
from_year = 2019
to_year = 2022
if run_extraction:
    fomc_potential_dates = deu.get_fomc_dates(from_year, to_year)

    # For each potential date, we now try to extract the minutes, statements, and press conference transcript
    statements, minutes, press_conferences = [], [], []

    fomc_dates = []
    for date in fomc_potential_dates:
        # Minutes and Statements are better formatted in HTML than pdf
        statements.append(deu.get_fomc_text(str(date), "statement", type='html'))
        minutes.append(deu.get_fomc_text(str(date), "minutes", type='html'))
        press_conferences.append(deu.get_fomc_text(str(date), "press"))
        fomc_dates.append(dt.datetime.strptime(date, "%Y%m%d"))
    
    fomc_data = {}
    fomc_data["dates"] = fomc_dates
    fomc_data["statements"] = statements
    fomc_data["minutes"] = minutes
    fomc_data["press"] = press_conferences
    
    with open('data/fomc_data_{}.pkl'.format(str(dt.date.today())), 'wb') as handle:
        pickle.dump(fomc_data, handle, protocol=pickle.HIGHEST_PROTOCOL)
    
else:
    if not pickle_file_name:
        raise ValueError("Please provide extraction target")
        
    with open(pickle_file_name, "rb") as input_file:
        fomc_data = pickle.load(input_file)
        

    
```

## Modelling

In this model, we will use LSTM with pre-trained GloVe embedding vectors.

```python
data_x, data_y = deu.get_clean_train_data(agg_data, directions)
```

```python
glove_file = 'glove.6B.300d.pickle'
glove_path = "data/" + glove_file

load = False

if load:
    embedding_dict = {}

    with open('data/' + "glove.6B.300d.txt", 'r', encoding="utf8") as f:
        for line in f:
            values = line.split()
            word = values[0]
            vectors = np.asarray(values[1:], 'float32')
            embedding_dict[word] = vectors
    f.close()

    pickle.dump(embedding_dict, open(glove_path, 'wb'))
    
glove_dict = pickle.load(open(glove_path, 'rb'))

```

```python
batch_size = 32
EMBEDDING_DIM = 300
epochs = 5
tokenizer = Tokenizer(num_words=200, split=' ') 
tokenizer.fit_on_texts([" ".join(x) for x in data_x])
word_index = tokenizer.word_index

embedding_matrix = np.zeros((len(word_index) + 1, EMBEDDING_DIM))
for word, i in word_index.items():
    embedding_vector = glove_dict.get(word)
    if embedding_vector is not None:
        # words not found in embedding index will be all-zeros.
        embedding_matrix[i] = embedding_vector

X = tokenizer.texts_to_sequences([" ".join(x) for x in data_x])
X = pad_sequences(X)
train_x, test_x, train_y, test_y = train_test_split(X, pd.get_dummies(data_y).to_numpy(), test_size=0.10, random_state=42)
   
```

```python
model = Sequential()
model.add(Embedding(len(word_index)+1, EMBEDDING_DIM,
    embeddings_initializer=keras.initializers.Constant(embedding_matrix),
    trainable=False,))
model.add(SpatialDropout1D(0.4))
model.add(LSTM(32, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(3,activation='softmax'))
model.compile(loss = 'categorical_crossentropy', optimizer='adam', metrics = ['accuracy'])
print(model.summary())
```

```python
epochs = 50
history = model.fit(train_x, train_y, validation_split = 0.1, 
                    epochs = epochs, batch_size=batch_size,
                    shuffle=True, verbose = 'auto')
test_loss, test_accuracy = model.evaluate(test_x,test_y)

```

```python
fig, ax = plt.subplots()
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
ax.hlines(y=test_accuracy, xmin=0, xmax=50, linewidth=2, color='r')
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
prediction = model.predict(test_x)
prediction = np.argmax(prediction, axis=1)
results = list(map(lambda x: x* (-1) + 1, prediction))
actual = np.argmax(test_y, axis=1)
actual = list(map(lambda x: x* (-1) + 1, actual))
cm = confusion_matrix(actual, results)
total = sum([sum(l) for l in cm])
cm = [[round(x / total, 3) for x in l] for l in cm]
class_names = [ 'Raise', 'Hold', 'Lower']
f = sns.heatmap(cm, annot=True, xticklabels=class_names, yticklabels=class_names, cmap="YlGnBu")
```
